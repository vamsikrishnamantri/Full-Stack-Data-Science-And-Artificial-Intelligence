```python
#lmplot(): it is a predefined function in seaborn library.
#it is used to visualize the bivariate analysis(i.e by using two variable/attributes).
#here in lmplot, lm means linear model.
#it has data,x,y.....etc parameters. We need to pass the dataframe to the data parameter, two attribiutes for x and y parameters respectivly
%matplotlib inline
plt.rcParams['figure.figsize']=6,2
```

```python
vis=sns.lmplot(data=bank_data[(bank_data.Geography=='France') & (bank_data.Gender=='Female')],y='EstimatedSalary',x='Age')#data of france female age and creditScores.  
vis.figure
```

```python
#fit_reg is a paremeter in lmplot(), it is used to manage the linear line in the graph, it bydefault it is Ture by parameter tunning.
#if we not need that line then we have to perform the hyper parameter tunning.
vis=sns.lmplot(data=bank_data[(bank_data.Geography=='France') & (bank_data.Gender=='Female')],y='EstimatedSalary',x='Age',fit_reg=False)
vis.figure
```

```python
#the equation of the linear line is y=mx+c.
#if the degree of x increases, then the curve of the line increases.
```

```python
vis=sns.lmplot(data=bank_data[(bank_data.Geography=='France')],y='EstimatedSalary',x='Age',fit_reg=True, hue='Gender')
vis.figure
```
