```python
#if elif else:
# --> Requirement is if we give 1 then we need output as 'one'. if we give 2 then output as 'two'. if we give 3 then the output is 'three'
```

```python
#if we go with if statement:

x=1
if x==1:
    print('one')
if x==2:
    print('two')
if x==3:
    print('three')
```

```python
#in the above program our requirement is satisfied at 1st block only but our machine will execute the remaining code also.
# it is waste of execution time and memory.
# so to avoid this condition, we need to use if elif else statement.
```

```python
#if we go with if statement:

x=1
if x==1:
    print('one')
elif x==2:
    print('two')
elif x==3:
    print('three')
```

```python
#in the above program, our requirment is satisfied at 1st block. The remaining blocks will not execute by the machine.
#So we saved our time and memory.
```
