```python
#Task:- Convert Raw Data To Clean Data Using Python split() function.

#we are having a raw dataset, we need to convert in as a clean dataset.
```

```python
raw_data=pd.read_excel(r"D:\Temp\Data Science & Artificial Intelligence\Completed\Week-3\Class-15 on 06-02-2024 on Tuesday\Notes\6th\6th\TASK -- convert raw data - clean data.xlsx")
```

```python
#This is our raw dataset
#in row 2,3 we have raw data.
#we need to convert it as clean data as in row 7,8.
raw_data
```

```python
#here i'm taking the raw to create a new dataframe.
#in excel it is dataset, in database it is table, in pandas it is dataframe.
raw=pd.DataFrame(raw_data.loc[2:3,'Unnamed: 2':'Unnamed: 5'])
#pd,DataFrame() is used to create a new dataframe.
#.loc[] & iloc[] are used to extract specific rows and columns in a dataframe.
#as of now don't worry about loc[] and iloc[].

raw.columns=raw_data.loc[1:1,'Unnamed: 2':'Unnamed: 5'].values.tolist()[0]
#in above line i'm changing the raw dataframe column names
```

```python
clean=pd.DataFrame(raw_data.loc[4:6,'Unnamed: 2':])
#here creating new dataframe as clean.
```

```python
clean.columns=raw_data.loc[7:7,'Unnamed: 2':].values.tolist()[0]
#in above line i'm changing the clean dataframe column names
```

```python
#now we are having raw and clean dataframes
```

```python
raw
```

```python
clean
```

```python
#here i'm reseting the index of the dataframes
raw=raw.reset_index().drop('index',axis=1)
clean=clean.reset_index().drop('index',axis=1)
```

```python
clean.drop(2,axis=0,inplace=True)
```

```python
clean
```

```python
#This loop will convert the raw data into clean data
for i in range(len(raw)):
    clean.iloc[i:i+1,0:4]=str(raw.loc[i:i].values.tolist()[0][0]).split(',')
    clean.iloc[i:i+1,3:5]=str(clean.loc[i:i,'state':'state'].values.tolist()[0][0]).split()
    clean.iloc[i:i+1,5:8]=str(raw.iloc[i:i+1,1:].values.tolist()[0][0]).split()[0].split('-')[::-1]
    clean.iloc[i:i+1,8:9]=raw.iloc[i:i+1,2:3].values.tolist()[0]
    clean.iloc[i:i+1,9:]=str(raw.iloc[i:i+1,3:].values.tolist()[0][0]).split()[0].split(':')
```

```python
clean#this is the clean data.
```
