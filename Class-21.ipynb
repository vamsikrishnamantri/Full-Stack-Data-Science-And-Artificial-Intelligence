```python
"""
=>till now we learned about univariate and bivariate analysis.
=>now it's time to visualize the graphs by using multiple variables.
=>We can visualize multiviarate analysis by using Facetgrid(), subplots(), heatmap() in seaborn
"""
```

```python
#FacetGrid() is used to plot the graphs based on multiple variables/attributes.
g=sns.FacetGrid(movies_data,row='Genre',col='Year',hue='Year')
#map() is used to appling the ploting function to every graph.
g=g.map(plt.scatter,'CriticRating','AudienceRating')
plt.show()
```
