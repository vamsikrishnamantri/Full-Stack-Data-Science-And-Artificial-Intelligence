```python
# print the values divisible by 2

for i in range(10):
    if i%2==0:
        print(i)
```

```python
# print the vaules which are not divisible by 2

for i in range(10):
    if i%2!=0:
        print(i)
```

```python
# We are having thre statement : 1-break, 2-continue, 3-pass
```

```python
# break: it is used to break a loop/function and come out from it.

for i in range(5):# here the list is not completed but we use break when the i value is 2. that's why the loop breaks wheneven i value is 2.
    print(i)
    if i==2:
        break
```

```python
# continue: it is used to skip an iteration or some steps at some a specific point without breaking the loop.

for i in range(10): # here i don't want to print number 5
    if i==5:
        continue
    print(i)# observe the output.
```

```python
# print the values form 1 to 50. But don't print the values divisible by 3 or 5

for i in range(1,50):
    if i%3!=0 and i%5!=0:
        print(i)
```
