```python
#this function will display the plot of multiple players data.
#we need to pass the typ=(Salary/Points/Games) and plrs=(0/2/3......8/9)
def plotting(typ,plrs):
    for i in plrs:
        plt.plot(typ[i],label=Players[i])
        plt.xticks(ticks=range(10),labels=Seasons)#it is used to display the labels in x-axis 
        plt.legend()
    plt.show()
```

```python
plotting(Games,[0,1,2,5,3])
```

```python
plotting(Salary,[0,5])
```
